```python
def show_instructions():
    """Display the game instructions."""
    print("""
    Welcome to the Adventure Game!
    Commands:
        go [direction]
        get [item]
    """)

def show_status():
    """Display the player's current status."""
    print("---------------------------")
    print(f"You are in the {current_room}")
    print(f"Inventory: {inventory}")
    if "item" in rooms[current_room]:
        print(f"You see a {rooms[current_room]['item']}")
    print("---------------------------")

# A dictionary linking a room to other rooms
rooms = {
    'Hall': {
        'south': 'Kitchen',
        'east': 'Dining Room',
        'item': 'key'
    },
    'Kitchen': {
        'north': 'Hall',
        'item': 'monster'
    },
    'Dining Room': {
        'west': 'Hall',
        'south': 'Garden',
        'item': 'potion'
    },
    'Garden': {
        'north': 'Dining Room'
    }
}

# Start the player in the Hall
current_room = 'Hall'

# An inventory, which is initially empty
inventory = []

# Show the game instructions
show_instructions()

# Loop until the player quits
while True:
    # Show the current status
    show_status()

    # Get the player's next action
    move = input("> ").lower().split()

    # If the player types 'go' followed by a direction
    if move[0] == 'go':
        # Check if the move is valid (i.e., there's a room in that direction)
        if move[1] in rooms[current_room]:
            current_room = rooms[current_room][move[1]]
        else:
            print("You can't go that way!")

    # If the player types 'get' followed by an item
    elif move[0] == 'get':
        # If the room contains an item and it matches the player's input
        if 'item' in rooms[current_room] and move[1] == rooms[current_room]['item']:
            # Add the item to the inventory
            inventory.append(move[1])
            # Remove the item from the room
            print(f"{move[1]} got!")
            del rooms[current_room]['item']
        else:
            print(f"Can't get {move[1]}!")

    # If the player types 'quit'
    elif move[0] == 'quit':
        break

    # Check if the player has encountered the monster
    if 'item' in rooms[current_room] and rooms[current_room]['item'] == 'monster':
        print("A monster has got you... GAME OVER!")
        break

    # Check if the player has won
    if current_room == 'Garden' and 'key' in inventory and 'potion' in inventory:
        print("You escaped the house with the key and potion... YOU WIN!")
        break
```
